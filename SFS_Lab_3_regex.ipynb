10. Use regex to find any ip addresses?

11. Create a script that will ask for user input, ask the user to enter a letter, then use regex to return any matches of that letter in the file?

```python
# your code here
import re

user_letter = input("Enter a letter: ")

with open("apache_log", "r") as file:

    for line in file:
        if re.search(user_letter , line):
            print(line)
    else:
        print("The is no match")
```
